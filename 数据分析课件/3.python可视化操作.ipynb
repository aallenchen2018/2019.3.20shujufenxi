### 1. 基本用法

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False 
```

```python
x=np.arange(0,4,0.01)
y=np.sin(2*np.pi*x)
plt.figure(figsize=(10,6))#改变图片大小
plt.plot(x,y,'g-.',linewidth=1)#绘制线状图
plt.xlim(0,4)#改变x轴的显示区域
plt.ylim(-1,1)#改变y轴的显示区域
plt.grid()#显示网格
plt.xlabel('时间(s)',fontsize=12)#x轴的标签
plt.ylabel('电压(V)',fontsize=12)#y轴的标签
plt.title('电压变化趋势图',fontsize=15)
plt.text(1.5,0.25,'y=sin(2*pi*x)',fontsize=12)#插入文本\
plt.text(1,0.75,'学个锤子',fontsize=30)
# plt.show()#显示图片
plt.savefig('./image/Chuizidianya.png',dpi=400,bbox_inches='tight')#保存
```

### 2. 绘制子图

```python
import matplotlib.pyplot as plt 
import numpy as np
import pandas as pd
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False 
```

```python
fig = plt.figure(figsize=(10,15))
ax=fig.subplots(3,2)

#直方图
x = np.random.normal(5,10,(10000))
ax[0][0].hist(x,bins=50,alpha=0.5,density=True)
y = np.random.normal(0,1,(10000))
ax[0,0].hist(y,bins=50,alpha=0.5,color='g',density=True)

#散点图
x1 = np.random.randint(0,50,100)
y1 = np.random.randint(0,50,100)
area = (30 * np.random.rand(100))**2
ax[0,1].scatter(x1,y1,color=['r','g','b'],marker='*',s=area,alpha=0.5)

#线状图
x2 = np.random.randint(0,50,100)
y2 = np.sin(x2)
ax[1,0].plot(x2,y2,'r--',marker='s',markeredgecolor='g',markerfacecolor='g')

#柱状图
y3 = np.random.randint(20,100,5)
ax[1,1].bar(range(1,6),y3,color=['r','g','b'],alpha=0.5,width=0.5)

#饼状图
x=[28,25,23,20,4]
labels=['HUAWEI','XIAOMI','OPPO','VIVO','OTHERS']
explode = [0.05,0,0,0,0]
patches, texts, autotexts=ax[2,1].pie(x,explode=explode,labels=labels,autopct='%.1f%%',radius=1)
for text in texts:
    text.set_size(15)
for autotext in autotexts:
    autotext.set_size(12)
plt.savefig('image/Chedantu.png',dpi=300,bbox_inches='tight')
```

### 3. pandas的绘图功能

```python
data=pd.DataFrame(100*np.abs(np.random.rand(4,4)),index=['one','two','three','four'],columns=['HW','ZTE','ESSS','NOKIA'])
data
```

```python
#线状图
data.plot(kind='line',figsize=(10,6),grid=True,ylim=(0,200),xticks=range(0,4),xlim=(0,3))
```

```python
#柱状图
fig = plt.figure(figsize=(10,10))
ax = fig.subplots(2) 
data['ZTE'].plot(kind='bar',ax=ax[0])
ax[0].set_xticklabels(labels=data['ZTE'].index,rotation=0)
# plt.xticks(rotation=30)
data['ZTE'].plot(kind='barh',ax=ax[1])
```

```python
#柱状图
fig1 = plt.figure(figsize=(10,10))
ax1 = fig1.subplots(1,2) 
data.plot(kind='bar',ax=ax1[0])
data.T.plot(kind='bar',ax=ax1[1],stacked=True)#堆叠柱状图
```
